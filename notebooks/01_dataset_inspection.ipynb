```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/mental health.csv")
```

```python
df = df.dropna(subset=['statement'])
```

```python
df.isnull().sum()
```

```python
df['status'].value_counts().plot(kind='bar', color='skyblue')

plt.title('Distribution of Mental Health Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
label_mapping = {

'Normal': 0,

'Stress': 1,

'Anxiety': 1,

'Depression': 2,

'Suicidal': 2,

'Bipolar': 2,

'Personality disorder': 2

}

df['anxiety_level'] = df['status'].map(label_mapping)
```

```python
df['anxiety_level'].isnull().sum()
```

```python
df['anxiety_level'].value_counts()
```

```python
df['anxiety_level'].value_counts().plot(kind='bar', color='skyblue')

plt.title(' new Distribution of Mental Health Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
